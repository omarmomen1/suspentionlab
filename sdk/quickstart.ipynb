# SuspensionLab PRO SDK - Quickstart

```
from suspensionlab.client import SuspensionLabClient

# Initialize the client
client = SuspensionLabClient(api_key="sk_your_api_key_here")
```

## Run a Monte Carlo Sweep

```
params = {
    "m_s": 350.0,
    "m_u": 40.0,
    "k_s": 25000.0,
    "c": 2050.0,
    "k_t": 200000.0,
    "MR": 0.8
}

profile = {
    "profile_type": "iso8608",
    "iso_class": "B",
    "duration": 10.0
}

# Run stochastic sensitivity sweep
res = client.run_monte_carlo(params, profile, iterations=500, tolerance_pct=5.0)
print("Mean ISO Comfort:", res['result']['mean_rms_body_accel_wk'])
```
